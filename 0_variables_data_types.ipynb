# Variable and Data Types

One of the things we use technology for the most is to store data for us. You will be used to having files on your computer, and these files will be of different types e.g. .txt, .png, .mp4

When we are writing code, we often want to store values in our program. To do this we can store them in a variable. 

## What is a variable?
A variable is a named space in the computer's memory that can hold a piece of data. (This is a bit of a simplification, but for now this is all we need.)

When we create a variable - meaning we tell the computer to reserve a bit of memory and give it a name - we say we are **declaring** the variable. 

When we put a value in our variable - we say we are **assinging** the value to the variable.
